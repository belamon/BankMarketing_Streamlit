```python
import matplotlib.pyplot as plt

# Data for the donut chart
labels = [
    "True Positives (Interested, Correctly Predicted)", 
    "False Positives (Uninterested, Incorrectly Predicted)", 
    "True Negatives (Uninterested, Correctly Predicted)", 
    "False Negatives (Interested, Missed)"
]
sizes = [77, 17, 83, 23]  # Corresponding values
colors = ["#4CAF50", "#FF5722", "#2196F3", "#FFC107"]  # Colors for each category

# Create the donut chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=colors, 
    textprops={'fontsize': 10}, 
    wedgeprops={'width': 0.4}  # Create the donut effect
)

# Customize the title
ax.set_title("Customer Classification with Model", fontsize=16)

# Show the chart
plt.show()
```
